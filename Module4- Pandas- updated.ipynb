# Data Manipulation

- Pandas is a powerful and easy-to-use open source data manipulation and analysis library in Python. 
- It is designed to make it easier to work with structured data (such as data in a CSV file or SQL database), and to perform operations such as slicing, grouping, and summarizing data, as well as preparing data for further analysis and visualization.
- It is widely used in data science, machine learning, and scientific computing.

## Pandas and its Features

####  1.Fast and efficient data manipulation and processing:
- Pandas is designed to make it easy and efficient to manipulate and analyze large datasets. It includes a wide range of functions and methods for filtering, grouping, aggregating, and transforming data, as well as handling missing values and time series data.

#### 2.Easy handling of missing data (represented as NaN in Pandas): 
- Pandas provides various functions and methods for handling missing data, such as filling missing values with a given value or interpolating missing values based on the data.


#### 3.Allows for easy grouping and aggregation of data: 
- Pandas provides the groupby function and the pivot_table and crosstab methods for grouping and aggregating data. You can use these functions to calculate statistics, apply functions to groups of data, and reshape and pivot the data.


#### 4.Provides robust input/output capabilities for a variety of file formats (e.g. CSV, Excel, etc.) and databases:
- Pandas can read and write data to and from a variety of file formats (e.g. CSV, Excel, JSON, HDF5, etc.) and databases (e.g. SQL, MySQL, etc.). It makes it easy to load and save data for analysis and modeling.


#### 5.Flexible reshaping and pivoting of data: 
- Pandas provides functions and methods for reshaping and pivoting data, such as pivot, pivot_table, and stack/unstack. These can be useful for transforming and cleaning data, as well as for preparing data for visualization.


#### 6.Time series functionality:
- Pandas has robust support for working with time series data. It includes functions for handling date and time data, resampling and frequency conversion, and moving window calculations.


#### 7.Merging, joining, and concatenating of data: 
- Pandas provides functions for merging and joining data from multiple sources, as well as concatenating data along different axes. These can be useful for combining data from different sources, or for reshaping and pivoting data.


#### 8.Provides many options for data visualization: 
- Pandas can be used together with Matplotlib and other data visualization libraries to create a wide range of plots and charts, such as line plots, scatter plots, bar plots, histograms, etc.

# Data Structures of pandas
- There are two main data structures in Pandas: the Series and the DataFrame.

### Series
- A Series is a one-dimensional labeled array that can hold any data type. 
- It is similar to a column in a spreadsheet or a field in a database table. 
- A Series has a name (the label) and an index (the row labels), and the data is stored in an NumPy array.

#### Visual representation of a Pandas Series:

  index   |   data
----------|----------
   0      |   value1
   1      |   value2
   2      |   value3
   ...    |   ...     

### DataFrame
- A DataFrame is a two-dimensional labeled data structure with rows and columns. 
- It is similar to a spreadsheet or a SQL table. 
- A DataFrame can have many columns, each with a different data type (e.g. string, integer, float, etc.).
- A DataFrame has a row index and a column index, and the data is stored in a NumPy array.

#### Visual representation of a Pandas DataFrame:

 index    |  column1  |  column2  |  column3  |  ...
----------|-----------|-----------|-----------|-----------
   0      |   value1  |   value2  |   value3  |   ...
   1      |   value4  |   value5  |   value6  |   ...
   2      |   value7  |   value8  |   value9  |   ...
   ...    |   ...     |   ...     |   ...     |   ...

## Join method – index to index
- The join method aligns two Pandas objects based on common index values. 
- That is,it looks for matching index values in both objects and then align them vertically. 
- The default type of join for this method is a left join.
- Let us consider the following example where we join two objects.

````python
periodic_table=pd.DataFrame({'Atomic Number':[1,2,3,4,5],'Element':['Hydrogen','Helium','Lithium','Beryllium','Boron'],'Symbol':['H','He',
'Li','Be','B']})
periodic_table
````

````python
periodic_table_extended=pd.DataFrame({'Atomic Number':[8,9,10],'Element':['Oxygen','Fluorine','Neon'],'Symbol':['O','F','Ne']},index=[1,20,30])
periodic_table_extended
````

````python
periodic_table.index  #{0,1,2,3,4}
````

````python
periodic_table_extended.index #{0,1,2}
````

````python
#Join columns of another DataFrame.
````

````python
periodic_table.join(periodic_table_extended,lsuffix='_Left', rsuffix='_Right',how='left')  #default 'left'
````

````python
periodic_table.join(periodic_table_extended,lsuffix='_Left', rsuffix='_Right',how='right')  #default 'left'
````

````python
periodic_table.join(periodic_table_extended,lsuffix='_Left', rsuffix='_Right',how='outer')  #default 'left'
````

````python
periodic_table.join(periodic_table_extended,lsuffix='_Left', rsuffix='_Right',how='inner') 
````

````python
periodic_table.join(periodic_table_extended,lsuffix='_Left', rsuffix='_Right',how='cross') 
````

### Merge method – SQL type join based on common columns
- Like the join method, the merge method is also used to join objects horizontally. 
- It is used when we join two DataFrame objects with a common column name. 
- The maindifference between the join and merge methods is that
- the join method combines the objects based on common index values, while the merge method combines the objects based on common column names. 
- Another difference is that the default join type in the merge method is an inner join, 
- while the join method performs a left join of the objects by default.
- Let us look at how the merge method works with an example. The two DataFrame objects defined here have a column name that is common – Atomic Number.
- This is a scenario where we can apply the merge method.

````python
periodic_table=pd.DataFrame({'Atomic Number':[1,2,3,4,5],'Element':['Hydrogen','Helium','Lithium','Beryllium','Boron'],
                             'Symbol':['H','He','Li','Be','B']})
periodic_table
````

````python
periodic_table_extended=pd.DataFrame({'Number':[1,2,3],'Natural':'Yes'})
periodic_table_extended
````

````python
#Merge DataFrame or named Series objects with a database-style join.
````

````python
periodic_table.merge(periodic_table_extended,how='inner',left_on='Atomic Number',right_on='Number')
````

### Tidy data and techniques for restructuring data
- Tidy data is a term developed by Hadley Wickham. According to a paper authored by him (Link: http://vita.had.co.nz/papers/tidy-data.pdf), these are the three principles of tidy data:
- 1. Columns correspond to variables in the data, and each variable maps to a single column.
- 2. The rows contain only observations, not variables.
- 3. Each data structure or table contains only one observational unit.
- Note that making data tidy is different from data cleansing. 
- Data cleansing is concerned with dealing with missing values and redundant information, removing filler characters, and changing inaccurate data types. 
- On the other hand, converting data to a tidy format involves restructuring the data and arranging it along the right axis, to facilitate ease of analysis.

### Imputation
- Imputation is the process of replacing missing values. 
- In Pandas, there is an option to substitute the missing values with measures of central tendencies like the mean, median,or mode, using the fillna method. You can also fill the missing values with a fixed or constant value like 0.
- We can use the forward fill technique to fill the missing value with the value just before it,or the backward fill technique to substitute the null value with the value just after it.

### forward fill - pad / ffill
- Let us substitute the NaN value for the country "Anguilla" with the values preceding it (in the previous row), using the forward fill technique (ffill), as shown in the following.

## Scatter plot
- Scatter plots draw data points using Cartesian coordinates to show the values of numerical values. 
- They also represent the relationship between two numerial values. 
- We can create a scatter plot in Matplotlib using the scatter() function, as follows:

````python
# Add the essential library matplotlib
import matplotlib.pyplot as plt

# create the data
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]

# Draw the scatter chart
plt.scatter(x, y, c='g', marker='o',alpha=0.8)

# Append the label on X-axis
plt.xlabel('X-label')

# Append the label on X-axis
plt.ylabel('Y-label')

# Add the title to graph
plt.title('Scatter Chart Sample')

# Display the chart
plt.show()
````

## Pie plot
- A pie plot is a circular graph that is split up into wedge-shaped pieces. Each piece is proportionate to the value it represents. The total value of the pie is 100 percent:

````python
# Add the essential library matplotlib
import matplotlib.pyplot as plt

# create the data
subjects = ['Mathematics','Science','Communication Skills','Computer Application']
scores = [85,62,57,92]

# Plot the pie plot
plt.pie(scores,labels=subjects,
        colors=['r','g','b','y'],
        startangle=0,
        counterclock=False,
        shadow= True,
        explode=(0,0.2,0,0),
        autopct='%1.1f%%')

# Add title to graph
plt.title('Student Performance')
# Add legend to graph
plt.legend(loc=(1.02,0))

# Draw the chart
plt.show()
````

## Bar plot
- A bar plot is a visual tool to compare the values of various groups. It can be drawn horizontally or vertically. We can create a bar graph using the bar() function:

````python
# Add the essential library matplotlib
import matplotlib.pyplot as plt

# create the data
movie_ratings = [1,2,3,4,5]
rating_counts = [21,45,72,89,42]

# Plot the data
plt.bar(movie_ratings, rating_counts, color='blue')

# Add X Label on X-axis
plt.xlabel('Movie Ratings')

# Add X Label on X-axis
plt.ylabel('Rating Frequency')

# Add a title to graph
plt.title('Movie Rating Distribution')
# Show the plot
plt.show()
````

## Histogram plot
- A histogram shows the distribution of a numeric variable. We create a histogram using the hist() method.
- It shows the probability distribution of a continuous variable. A histogram only works on a single variable while a bar graph works on two variables:

````python
# Add the essential library
import matplotlib.pyplot as plt

# Create the data
employee_age = [21,25,25,26,28,30,32,34,35,35,37,42,47,55]

# Create bins for histogram
bins = [20,30,40,50,60]

# Plot the histogram
plt.hist(employee_age,bins,  rwidth=0.6)

# Add X Label on X-axis
plt.xlabel('Employee Age')

# Add X Label on X-axis
plt.ylabel('Frequency')

# Add title to graph
plt.title('Employee Age Distribution')

# Show the plot
plt.show()
````

## Bubble plot
- A bubble plot is a type of scatter plot. It not only draws data points using Cartesian coordinates but also creates bubbles on data points. 
- Bubble shows the third dimension of a plot. It shows three numerical values: two values are on the x and y axes and the third one is the size of data points (or bubbles):

````python
import matplotlib.pyplot as plt
import numpy as np

# Set figure size
plt.figure(figsize=(8,5))

# Create the data
countries = ['Qatar','Luxembourg','Singapore','Brunei','Ireland','Norway','UAE','Kuwait']
populations = [2781682, 604245,5757499,428963,4818690,5337962,9630959,4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356,69382, 67000]

# scale GDP per capita income to shoot the bubbles in the graph
scaled_gdp_per_capita = np.divide(gdp_per_capita, 80)

colors = np.random.rand(8)

# Draw the scatter diagram
plt.scatter(countries, populations, s=scaled_gdp_per_capita, c=colors, cmap='Blues',edgecolors='grey', alpha=0.5)

# Add X Label on X-axis
plt.xlabel('Countries')

# Add Y Label on X-axis
plt.ylabel('Population')

# Add title to graph
plt.title('Bubble Chart')

# rotate x label for clear visualization
plt.xticks(rotation=45)

# Show the plot
plt.show()
````
